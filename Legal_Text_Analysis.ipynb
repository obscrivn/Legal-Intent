```python
df = pd.read_csv('/Users/KristinDay/Dropbox/LEGAL-DOMAIN-WORK/Pragmatics-paper/Current_Data/IN-case-paragraphs.csv')
```

```python
cases = []
for idx,row in df.iterrows():
    df_case = pd.DataFrame()
    df_case['Paragraphs'] = row
    df_case['Tokenize'] = ''
    df_case['POS_tag'] = ''
    df_case['Syntax'] = ''
    df_case.drop(df_case.index[:1],inplace=True)
    cases.append(df_case)
```

```python
def tokenize(x):
    if x != '':
        try:
            text = nltk.word_tokenize(x)
        except:
            text = None
    
    return text
```

```python
def pos_tag(x):
    if x != '':
        try:
            pos = nltk.pos_tag(x)
        except:
            pos = None
    return pos
```

```python
def clean_p(x):
    if type(x) == str:
        x = re.sub(r'\(.*?\)', '', x)
        #x = re.sub(r'\.', ' ', x)
        x = re.sub(r',,', ', ', x)
        #x = re.sub(r'!', ' ', x)
        #x = re.sub(r'\?', ' ', x)
    return x

for case in cases:
    case['Paragraphs'] = case['Paragraphs'].apply(lambda x: clean_p(x))
```

```python
for case in cases:
    case.replace(to_replace='NaN', value='', inplace=True)
    case['Tokenize'] = case['Paragraphs'].apply(lambda x: tokenize(x))
    case['POS_tag'] = case['Tokenize'].apply(lambda x: pos_tag(x))
```

```python
def get_string(x):
    if x != None:
        a_list = ''
        for i in range(0,len(x)):
            a_list += x[i][0] + '\t' + x[i][1] + '\n'
    else: a_list = []
    return a_list
```

```python
for case in cases:
    case['string'] = case['POS_tag'].apply(lambda x: get_string(x))
```
